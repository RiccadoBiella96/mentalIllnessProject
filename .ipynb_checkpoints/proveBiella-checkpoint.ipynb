## Analisi dei lavoratori in proprio e non e della ricerca di una cura per una malattia mentale

#### Cerchiamo una correlazione tra il fatto di lavorare in proprio e il fatto di aver cercato una cura per una malattia mentale 

Notiamo che la variabile "self_employed" assume spesso il valore nan (il campo non è stato compilato dall'utente). Creiamo quindi un nuovo dataset dove eliminiamo le righe (nel nostro caso gli utenti) del dataframe dove il campo "self_employed" non è stato compilato. Nel seguito lavoreremo sul nuovo dataset. 

Raggruppiamo la variabile "self_employed" nei seguenti 2 gruppi: 
- Primo gruppo di nome not self employed (questo gruppo identifica tutti i lavoratori dipendenti: tutti quelli che hanno idicato {no} nel campo "self_employed") 
- Secondo gruppo di nome self employed (questo gruppo identifica tutti i lavoratori in proprio: quelli che NON hanno indicato 'no' nel campo 'self_employed')

Raggruppiamo la variabile "tratment" nei seguenti 2 gruppi: 
- Primo gruppo di nome no treatment (questo gruppo identifica tutte le persone che hanno cercato una cura per una malattia mentale: tutti quelli che hanno idicato {no} nel campo "treatment") 
- Secondo gruppo di nome treatment (questo gruppo identifica tutte le persone che hanno cercato una cura per una malattia mentale: quelli che NON hanno indicato 'no' nel campo 'treatment')

Costruisciamo la tabella di contingenza per la coppia di variabili "self_employed" e "treatment" (usando i gruppi definiti sopra).

```python
#Creazione del nuovo dataFrame che elimina gli utenti che hanno indicato nan nel campo "self_employed"
dfnew = df.dropna(subset=['self_employed'])
self_employed = dfnew['self_employed'].values
#print(set(self_employed))


self_employedYesNo = np.full(self_employed.shape,"",dtype="object")
for i in range(len(self_employed)):
    if(self_employed[i]=='No'):
        self_employedYesNo[i] = "not self employed"
    else:
        self_employedYesNo[i] = "self employed"
#print(set(self_employedYesNo))
treatment = dfnew['treatment'].values
treatmentYesNo = np.full(treatment.shape,"",dtype="object")
for i in range(len(treatment)):
    if(treatment[i]=='No'):
        treatmentYesNo[i] = "no treatment"
    else:
        treatmentYesNo[i] = "treatment"
setSelf_employedYesNo = np.unique(self_employedYesNo)
setTreatmentdYesNo = np.unique(treatmentYesNo)
table = np.full((len(setSelf_employedYesNo), len(setTreatmentdYesNo)), np.nan, dtype=int)
for i in range (len(setSelf_employedYesNo)):
    for j in range (len(setTreatmentdYesNo)):
        table[i,j] = np.sum((self_employedYesNo==setSelf_employedYesNo[i]) & (treatmentYesNo==setTreatmentdYesNo[j]))
# pd.DataFrame(table, index=setSelf_employedYesNo, columns=setTreatmentdYesNo)

freq = np.sum(table, axis=1, keepdims=True) # somma i dipendenti e i lavoratori in proprio
people = freq[0]+freq[1]
print(f"\nNumber of people: {people[0]}")
print(f"\nNot self employed: {freq[0][0]} , {(freq[0]*100/people)[0]:.2f} %")
print(f"\nSelf employed: {freq[1][0]} , {(freq[1]*100/people)[0]:.2f} %")
t_cond_rows = table / freq
#print(t_cond_rows)
print(f"\nNot Self employed and no treatment: {t_cond_rows[0][0]*100:.2f} %")
print(f"\nNot Self employed and treatment: {t_cond_rows[0][1]*100:.2f} %")
print(f"\nSelf employed and no treatment: {t_cond_rows[1][0]*100:.2f} %")
print(f"\nSelf employed and treatment: {t_cond_rows[1][1]*100:.2f} %")
pd.DataFrame(table, index=setSelf_employedYesNo, columns=setTreatmentdYesNo)
```

L'istogramma generato al punto precedente che rappresenta la distribuzione delle età degli utenti disoccupati non è ben visibile nella stessa figura dell'istogramma che rappresenta la distribuzione delle età degli utenti lavoratori poichè il numero di utenti disoccupati è molto piu' basso rispetto ai lavoratori. Di conseguenza, anche le barre dell'istogramma sono molto piu' basse rispetto alle barre dell'istogramma caratterizzante gli utenti lavoratori. In questo caso, sarebbe meglio "normalizzare" gli istogrammi, dove l'altezza delle barre non è la frequenza assoluta (numero di utenti in un intervallo di età), ma la frequenta relativa (frazione di utenti in un intervallo di età). La cella sotto mostra come generare un istogramma normalizzato (basta impostare l'attributo ``histnorm`` uguale a ``probability'`` quando si creare l'oggetto della classe ``go.Histogram``).  

```python
#Hint
x = np.random.randn(1000)

trace = go.Histogram(x = x, histnorm = 'probability', marker = dict(color = 'rgb(255,0,0)'))
layout = go.Layout(title = 'Esempio di istogramma normalizzato', xaxis = dict(title = 'x'), yaxis = dict(title = 'frequenze relative'))
fig = go.Figure([trace], layout)
py.iplot(fig)
```

```python
# Solution
data = []
for val in support:
    num_g = age[jobArray == val] # tempi per le osservazioni corrispondenti
    trace = go.Histogram(x=num_g, histnorm = 'probability', name=val)
    data.append(trace)

layout = go.Layout(
    barmode='group',
    xaxis=dict(title="Età"),
    yaxis=dict(title="Frequenze Relative"),
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```
